```python
#Numpy problem 1

import numpy as np

def myvander(myarray,N,increasing):
    
    
    if increasing==False:
        
        d=[[i**(N-j-1) for j in range(0,N,1) ]for i in myarray]
        
        ns=np.array(d)
        
        print(ns)
        
    elif increasing==True:
        
        d2=[[i**(N-j) for j in range(N,0,-1)  ] for i in myarray]
        ns2=np.array(d2)
        print(ns2)
        
    else:
        
        try:
            
            raise Exception("add the boolean parameter True or false")
            
        except Exception as e:
            
            print(e)
            print(e.args)
    
        
myarray=np.array([3,4,5,6])

N=4
        
myvander(myarray,N,increasing=False)        

print("____deault vander_______")

print(np.vander(myarray,N,increasing=False))

```

```python
#numpy problem 2

import numpy as np

def movingwindow(myarray,window):
    
    
    i=0
    
    for j in range(0,len(myarray)-window+1,1):
        
        
        ns=myarray[i:window]
        
        print(np.mean(ns))
        
        i=i+1
        window=window+1
        
        
arr=np.array([3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150])

window=3

movingwindow(arr,window)
```

```python
#pandas sql problem 2

import sqlalchemy
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Numeric
from sqlalchemy.orm import sessionmaker
from sqlalchemy.sql import func
import pandas as pd
import sqlite3


mypath='https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'

names=["age","workclass","fnlwgt","education","educationnum","maritalstatus","occupation","relationship","race","sex","capitalgain","capitalloss","hoursperweek","nativecountry","income"]


engine = create_engine('sqlite:///:memory:', echo=False)
conn=engine.connect()

Base = declarative_base()



mydf=pd.read_csv(mypath,names=names,sep=",\s")

mydf

class Adults(Base):
    
    __tablename__='adults'
    
    id=Column(Integer,primary_key=True)
    age=Column(Integer)
    workclass=Column(String)
    fnlwgt=Column(String)
    education=Column(String)
    maritialstatus=Column(String)
    occupation=Column(String)
    
    def __repr__(self):
         return "<Adults(id='%s',age='%s', workclass='%s', fnlwgt='%s', education='%s')>" % (
                   self.id, self.age, self.workclass, self.fnlwgt,self.education)
        
        
sampledf=mydf[['age','workclass','fnlwgt','education']]

listToWrite = sampledf.to_dict(orient='records')

Base.metadata.create_all(engine)

table = sqlalchemy.Table('adults',Base.metadata, autoload=True)

DBSession = sessionmaker(bind=engine)

session = DBSession()

conn.execute(table.insert(), listToWrite)

session.commit()

session.close() 

#inserted adults



#update queries

firstadult = Adults(age='44',workclass='Public',fnlwgt='343591',education='HS-grad')
secondadult= Adults(age='53',workclass='Army',fnlwgt='346253',education='HS-grad')
session.add(firstadult)
session.add(secondadult)
session.commit()

check=session.query(Adults).filter_by(workclass='Army').first()
check2=session.query(Adults).filter_by(workclass='Public').first()
print("-----updated queries------")
print(check)
print(check2)


#deleting queries

obj1=session.query(Adults).filter_by(id='2').one()
obj2=session.query(Adults).filter_by(id='3').one()

print("------deleting objects-------")

session.delete(obj1)
session.delete(obj2)
session.commit()
all_adults= session.query(Adults).all()
print(obj1 in all_adults)
print(obj2 in all_adults)

print("-------filter queries-----")
obj3=session.query(Adults).filter_by(id='5').one()
obj4=session.query(Adults).filter_by(id='6').one()

print(obj3)
print(obj4)

print('-------function queries-------')

obj5=session.query(func.sum(Adults.age).label('totalage')).scalar()
obj6=session.query(func.avg(Adults.age).label('avgage')).scalar()

print(obj5)
print(obj6)
```
